```python
class Stack:
    def __init__(self):
        self.items = []
        
    def push(self, element):
        self.items.append(element)
        
    def pop(self):
        return self.items.pop()
    
    def __len__(self):
        return len(self.items)
    
    def peek(self):
        return self.items[len(self.items)-1]
    
    def is_empty(self):
        return self.items == []
```

```python
numbers = [1,2,3,4,5,6]

stack = Stack()

for number in numbers:
    stack.push(number)

out = []

while not stack.is_empty():
    element = stack.pop()
    out.append(element)
    
print(out)    
```
